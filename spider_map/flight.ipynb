```python
import pandas as pd
df1 = pd.read_csv('/Users/lizhengnan/desktop/storytelling quiz/flight/airlines.csv')
airports_df = pd.read_csv('/Users/lizhengnan/desktop/storytelling quiz/flight/airports.csv')
flights_df = pd.read_csv('/Users/lizhengnan/desktop/storytelling quiz/flight/flights.csv')
```

```python
airports_df.rename(columns={'IATA_CODE': 'AirportCode', 'LATITUDE': 'Latitude', 'LONGITUDE': 'Longitude'}, inplace=True)

# Merge origin airport latitude and longitude
flights_df = flights_df.merge(airports_df[['AirportCode', 'Latitude', 'Longitude']], left_on='ORIGIN_AIRPORT', right_on='AirportCode', how='left')
flights_df.rename(columns={'Latitude': 'Origin_Latitude', 'Longitude': 'Origin_Longitude'}, inplace=True)
flights_df.drop('AirportCode', axis=1, inplace=True)  # Remove the redundant column after merge

# Merge destination airport latitude and longitude
flights_df = flights_df.merge(airports_df[['AirportCode', 'Latitude', 'Longitude']], left_on='DESTINATION_AIRPORT', right_on='AirportCode', how='left')
flights_df.rename(columns={'Latitude': 'Dest_Latitude', 'Longitude': 'Dest_Longitude'}, inplace=True)
flights_df.drop('AirportCode', axis=1, inplace=True) 
flights_df.head()
```
